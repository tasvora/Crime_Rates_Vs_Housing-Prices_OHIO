```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
OHcrime_2017 = "Resources/OhioCrime2017.csv"
OHcrime_2016 = "Resources/OhioCrime2016.csv"
OHcrime_2015 = "Resources/OhioCrime2015.csv"
OHcrime_2014 = "Resources/OhioCrime2014.csv"
OHcrime_2013 = "Resources/OhioCrime2013.csv"
OHcrime_2012 = "Resources/OhioCrime2012.csv"

OHcrime_2017_df = pd.read_csv(OHcrime_2017)
OHcrime_2016_df = pd.read_csv(OHcrime_2016)
OHcrime_2015_df = pd.read_csv(OHcrime_2015)
OHcrime_2014_df = pd.read_csv(OHcrime_2014)
OHcrime_2013_df = pd.read_csv(OHcrime_2013)
OHcrime_2012_df = pd.read_csv(OHcrime_2012)

OHcrime2016_df = OHcrime_2016_df.dropna(how='any')
OHcrime2017_df = OHcrime_2017_df.dropna(how='any')
OHcrime2015_df = OHcrime_2015_df.dropna(how='any')
OHcrime2014_df = OHcrime_2014_df.dropna(how='any')
OHcrime2013_df = OHcrime_2013_df.dropna(how='any')
OHcrime2012_df = OHcrime_2012_df.dropna(how='any')
OHcrime2012_df.head()
```

```python
combined_crime_df = pd.merge(OHcrime_2017_df, OHcrime_2016_df, how = 'left', on="City")
combined_crime_df = combined_crime_df.dropna(how='any')
combined_crime_df.reset_index(drop=True)
#combined_crime_df.to_csv("Resources/CrimeData_merged.csv")
```

```python
combined_crime_df = combined_crime_df.rename(columns={"POPULATION_x":"2017 Population",
                                                              "VIOLENT CRIME_x":"2017 Violent Crime",
                                                              "MURDER_x":"2017 Murder", "RAPE_x": "2017 Rape",
                                                              "ROBBERY_x": "2017 Robbery", "AGGRAVATED ASSAULT_x": 
                                                              "2017 Aggravated Assault"})

# Rename our _y columns to 2016
combined_crime_df = combined_crime_df.rename(columns={"POPULATION_y":"2016 Population",
                                                              "VIOLENT CRIME_y":"2016 Violent Crime",
                                                              "MURDER_y":"2016 Murder", "RAPE_y": "2016 Rape",
                                                              "ROBBERY_y": "2016 Robbery", "AGGRAVATED ASSAULT_y": 
                                                              "2016 Aggravated Assault"})
# combined_crime_df.head()
```
